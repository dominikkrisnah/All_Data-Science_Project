```python
def krisna_linear(x,gradien):
    y = gradien*x
    return y
```

### Calculate the initial linear equation

```python
x_krisna = np.array([0,10])
m_awal = 5
y_krisna = krisna_linear(x_krisna,m_awal)
```

```python
plt.scatter(x,y)
plt.plot(x_krisna,y_krisna,"red")
plt.axis([0,12,0,12])
plt.show()
```

```python
m_prediksi = m_awal
m_list_prediksi = []
y_list_prediksi = []
x_list_prediksi = []
for i in range(1,data_size):
    y_krisna = krisna_linear(x[i],m_prediksi)
    y_actual = y[i]

    error = y_actual - y_krisna
    delta_m = error/x[i]
    m_prediksi = m_prediksi + delta_m

    m_list_prediksi.append(m_prediksi)
    y_list_prediksi.append(y_krisna)
    x_list_prediksi.append(x[i])
```
